# set 集合
1. 相當數學中的集合，比起list沒有順序。
2. 表示
 - 大括弧 {} 表示
 - 空集合不是空的大括弧，是set()
 
 
3. 集合的元素操作
 - 加入元素：set.add()
 - 刪除元素：set.remove()
 
 
4. 集合之間運算
 - 聯集 |
 - 交集 &
 - 差集 -
 - 聯集去掉交集 ^

```python
amy = {'singing', 'cooking', 'reading'}
ben = {'hiking', 'singing', 'cooking'}
chris = {'hiking', 'baking', 'swimming'}
```

```python
chris.remove('swimming')
chris.add('singing')
print(chris)
```

```python
amy_and_ben = amy | ben
print(amy_and_ben)
```

# dict 字典
 - 可視為list的延伸，用任意物件做為索引
 - 用大括號{ }表示
 - 每一項用key: value的方式表示，用逗號區隔
 

```python
david = {'name': 'david',
         'gender': 'male',
         'age': 35,
         'hobby': ['painting', 'hiking'],
         }
```

```python
print(david['hobby'])
```

```python
david['city'] = 'hualien'
print(david)
```

```python
david.pop('name')
print(david)
```

```python
david.setdefault('job', 'doctor')
print(david)
```
