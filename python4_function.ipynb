## 引數傳入的兩種方式
 - 位置引數(Positional Argument)：引數依照前後順序對應到參數
 - 關鍵字引數(Keyword Argument)：引數依照標籤名字對應到參數


兩種傳遞方式可以同時並用，規則是位置引數必須放在關鍵字引數前面。


```python
sentence = "This is a pencil."
words = sentence.split(" ")
print(words)
```

```python
words = sentence.split(" ", 1)
print(words)
```

```python
words = sentence.split(" ")
print(words)
```

```python
words = sentence.split(1)
print(words)
```

```python
words = sentence.split(maxsplit = 1)
print(words)
```

### 定義函數
#### 原則
 - 冒號 & 縮排
 - 以方便他人理解的方式命名
 - 以小寫命名

#### "return"
 - 回傳值即是函數物件的賦值
 - 當程式運行到 return 時將會無條件直接結束
 - 一個函式也可以選擇不回傳結果

```python
def hello(name):
    a = "你好，" + name + "！"
    print(a)
    
item = hello("周杰倫")
```

```python
# return
def hello(name):
    a = "你好，" + name + "！"
    return a
    
item = hello("周杰倫")
```

```python
def hello(name):
    a = "你好，" + name + "！"
    return a
    
item = hello("周杰倫")
print(item)
```

```python
# 程式運行到 return 時將會無條件直接結束
def hello(name):
    a = "你好，" + name + "！"
    return a
    a = a + "再見！"
    
item = hello("周杰倫")
print(item)
```

```python
def slope_of_triangle(a,b):
    sum_square = a ** 2 + b ** 2
    slope = sum_square ** 0.5
    return slope

print( slope_of_triangle(5,12) )
```

```python
# 全域變數與區域變數
print(sum_square)
```

```python
# 練習：製作梯形的面積公式
def area(upper, lower, height):
    return (upper + lower) * height / 2
    
    
```

## 可變參數函數 variadic functions
 - *args 
 - **kwargs
#### * 和 ** 在Python中視為一種「解包」語法

```python
x = 3
y = 6
z = 9
print(x, y, z)
```

```python
tup = (3,6,9)
print( tup )
```

```python
print(*tup)
```

```python
triangle_edge = (5, 12)
print( slope_of_triangle( triangle_edge ) )
```

```python
print( slope_of_triangle( *triangle_edge ) )
```

```python
param = {"end": "!!!", "sep": "!"}
print("Hello", "Tom", "Good morning", end = param["end"], sep = param["sep"])
```

```python
print("Hello", "Tom", "Good morning", **param)
```
