## Boolean type 布林型態
### 只有 True 和 False 兩個物件！
 - 任何其他物件都可以藉由bool()函數轉為對應的Boolean Type
 - 所有 bool() 值為 False 的物件：
  - 數值0
  - 空字串、空集合、空串列、空序對、空字典
  - 假的敘述
  - None
 - 其他物件的 bool() 值都為 True！

```python
a = true
print( type(a) )
```

```python
b = False
print( type(b) )
```

```python
number = 0
print( bool(number) )
```

```python
string = ""
print( bool(string) )
```

```python
string_2 = " "
print( bool(string_2) )
```

```python
new_list = []
print(new_list)
print( bool(new_list) )
```

```python
new_list.append(8)
print(new_list)
print( bool(new_list) )
```

## comparison operator 比較運算子
### 建立一個敘述，回傳True或False！
 - 總共六個
  - 小於/大於 < / >
  - 小於等於/大於等於 <= / >=
  - 「是否等於」 ==
  - 「是否不等於」 !=

```python
alice = 28
bob = 38
chris = 26
```

```python
statement = (alice == bob)
print(statement)
print( type(statement) )
```

## 迴圈：while 和 for...in...
### 目的：讓電腦重複做事

### while
 - while 後面接的是條件

```python
# 求出 30 的因數
number = 30
factor = 1

while factor <= int(number ** 0.5):
    if number % factor == 0:
        print(factor)
        print(int(number / factor))
    factor = factor + 1
```

```python
# 計算1至20的平方
square_list = []
number = 1
while number <= 20:
    square = number ** 2
    print(f"{number}的平方為{square}")
    square_list.append(square)
    number = number + 1
print(square_list)
```

```python
# 無限迴圈
square_list = []
number = 1
while number <= 20:
    square = number ** 2
    print(f"{number}的平方為{square}")
    square_list.append(square)

print(square_list)
```

### 實例：96孔盤

 - 橫排為row，直排為column
 - 位置：A1 A2 A3...

```python
# 二維迴圈
row_list = ['A', 'B', 'C', 'D', 'E', 'F', 'G', 'H']
column_list = range(1, 13)

for row in row_list:
    for column in column_list:
        print(row + str(column))
```

```python
for column in column_list:
    for row in row_list:
        print(row + str(column))
```

```python
# 創建二維list
well_list = []
for row in row_list:
    new_row = []
    for column in column_list:
        new_row.append( row + str(column) )
    well_list.append(new_row)

print(well_list)
```

### break & continue
 - break: 結束整個迴圈
 - continue: 結束該輪迴圈，繼續下一輪

```python
# 計算1至20的平方
square_list = []
number = 0
while number <= 20:
 
   number = number + 1
   if number == 15:
        break
   square = number ** 2
   print(f"{number}的平方為{square}")
   square_list.append(square)

print(square_list)
```

```python
# 計算1至20的平方
square_list = []
number = 0
while number <= 20:
    number = number + 1
    if number == 15:
        continue
    square = number ** 2
    print(f"{number}的平方為{square}")
    square_list.append(square)

print(square_list)
```

### 進階： iterator 和 iterable


```python
from itertools import product

row_list = ['A', 'B', 'C', 'D', 'E', 'F', 'G', 'H']
column_list = range(1, 13)

for cell in product(row_list, column_list):
    print(cell)
```

```python
from itertools import product

row_list = ['A', 'B', 'C', 'D', 'E', 'F', 'G', 'H']
column_list = range(1, 13)

for cell in product(row_list, column_list):
    position = cell[0] + str(cell[1])
    print(position)
```

```python
from itertools import chain

for cell in chain(row_list, column_list):
    print(cell)
print(list(chain(row_list, column_list)))
```

```python
from itertools import permutations
students = ["amy", 'ben', 'cathy', 'david', 'emily', 'fred', 'grace', 'henry']
for leaders in permutations(students, 2):
    print(leaders)

print(len(list(permutations(students, 2))))
```

```python
from itertools import combinations
students = ["amy", 'ben', 'cathy', 'david', 'emily', 'fred', 'grace', 'henry']
for leader in combinations(students, 2):
    print(leader)
print(len(list(combinations(students, 2))))
```
